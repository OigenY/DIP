1. Зчитати та завантажити тестове чорно-біле та кольорове зображення.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
bw = cv2.imread("C:/Temp/1_8.png", cv2.IMREAD_GRAYSCALE)
rgb = cv2.imread("C:/Temp/rgb.png")
```

2. Відобразити чорно-біле та кольорове зображення.

```python
plt.figure(figsize=(5, 5))
plt.imshow(bw, cmap='gray')
plt.show();
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(rgb)
plt.show();
```

3. Перетворити зображення з колірної моделі BGR в колірні моделі RGB і HSV та відобразити.

```python
plt.figure(figsize=(5, 5))
plt.imshow(cv2.cvtColor(rgb, cv2.COLOR_BGR2RGB))
plt.show();
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(cv2.cvtColor(rgb, cv2.COLOR_BGR2HSV))
plt.show();
```

4. Окремо відобразити колірні канали для зображень із колірними моделями RGB та HSV.

```python
plt.figure(figsize=(10, 10))
img_rgb = cv2.cvtColor(rgb, cv2.COLOR_BGR2RGB)
plt.subplot(131);
R = np.zeros(img_rgb.shape, dtype='uint8')
R[:, :, 0] = img_rgb[:, :, 0]
plt.imshow(R)

plt.subplot(132);
G = np.zeros(img_rgb.shape, dtype='uint8')
G[:, :, 1] = img_rgb[:, :, 1]
plt.imshow(G)

plt.subplot(133);
B = np.zeros(img_rgb.shape, dtype='uint8')
B[:, :, 2] = img_rgb[:, :, 2]
plt.imshow(B)

plt.show()
```

```python
plt.figure(figsize=(10, 10))
img_hsv = cv2.cvtColor(rgb, cv2.COLOR_BGR2HSV)
plt.subplot(131);
R = np.zeros(img_hsv.shape, dtype='uint8')
R[:, :, 0] = img_hsv[:, :, 0]
plt.imshow(R)

plt.subplot(132);
G = np.zeros(img_hsv.shape, dtype='uint8')
G[:, :, 1] = img_hsv[:, :, 1]
plt.imshow(G)

plt.subplot(133);
B = np.zeros(img_hsv.shape, dtype='uint8')
B[:, :, 2] = img_hsv[:, :, 2]
plt.imshow(B)

plt.show()
```

5. Скопіювати область інтересу (Region of Interest) чорно-білого зображення, використовуючи наступні діапазони 0:img.shape[0]/2, 0:img.shape[1]/2.

```python
bw = bw[int(0):int(bw.shape[0]/2),int(0):int(bw.shape[1]/2)]
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(bw, cmap='gray')
plt.show();
```
